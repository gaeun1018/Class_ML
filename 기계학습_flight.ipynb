```python
import pandas as pd
```

```python
train = pd.read_csv('flight_train.csv')
test = pd.read_csv('flight_test.csv')
```

```python
print(train.shape, test.shape)
```

```python
print(train.head())
```

```python
print(test.isnull().sum())
```

```python
print(train['price'].describe())
```

```python
target = train.pop('price')
```

```python
cols = train.select_dtypes('object').columns
for col in cols:
  set_train = set(train[col])
  set_test = set(test[col])
  same = (set_train == set_test)
  if not same:
    print(col,"카테고리다름")
```

```python
train = train.drop('flight',axis=1)
test = test.drop('flight',axis=1)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(random_state=7)
rf.fit(X_tr,y_tr)
```

```python
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
score = root_mean_squared_error(y_val, pred)
```

```python
print(score)

pred = rf. predict(test)

submit = pd.DataFrame({'pred':pred})
```

```python
submit.to_csv('result.csv',index=False)
submit.head()
```
