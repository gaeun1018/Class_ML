```python
from sklearn import datasets
import matplotlib.pylab as plt
import numpy as np
from sklearn.linear_model import LinearRegression

diabetes_X, diabetes_y= datasets.load_diabetes(return_X_y=True)

print("diabetes_X.shape= ", diabetes_X.shape)
print("diabetes_y.shape= ", diabetes_y.shape)
```

```python
feature_names = datasets.load_diabetes().feature_names

print(feature_names)
```

```python
# 하나의 특징(BMI)만 추려내서 2차원 배열로 만든다. BMI 속성의 인덱스는 2
diabetes_X_new= diabetes_X[:, np.newaxis, 2]
print("diabetes_X_new.shape= ", diabetes_X_new.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes_X_new,
                                                    diabetes_y, test_size=0.1, random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()

regr.fit(X_train,y_train)

y_pred = regr.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test,y_test,color='black')
plt.plot(X_test, y_pred,color='blue', linewidth=3)
plt.show()
```
