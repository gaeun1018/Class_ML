```python
import pandas as pd
```

```python
train = pd.read_csv('score_train.csv')
test = pd.read_csv('score_test.csv')
```

```python
print(train.shape, test.shape)
```

```python
print(train.head())
```

```python
print(train.isnull().sum())
```

```python
print(test['Credit_Score'].value_counts())
```

```python
target = train.pop('Credit_Score')
```

```python
cols = train.select_dtypes('object').columns
for col in cols:
  set_train = set(train[col])
  set_test = set(test[col])
  same = (set_train == set_test)
  if not same:
    print(col,"카테고리다름")
```

```python
#원핫인코딩 실행
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=7)
rf.fit(X_tr,y_tr)
```

```python
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
```

```python
score = f1_score(y_val, pred, average='weighted')
```

```python
print(score)

pred = rf. predict(test)

submit = pd.DataFrame({'pred':pred})
```

```python
submit.to_csv('result.csv',index=False)
submit.head()
```
